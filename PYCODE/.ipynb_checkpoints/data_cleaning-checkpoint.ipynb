```python
import pandas as pd
```

```python
df= pd.read_csv(r"D:\PM_DATA\CSV_DATA\pressure.csv", index_col=[0])
```

```python
df
```

```python
lat = df['lat'].nunique()
lon = df['lon'].nunique()
```

```python
print(lat)
print(lon)
```

```python
date = df[(df['time'] >= '2015-02-01 00:00:00') & (df['time'] <= '2023-08-31 18:00:00')]
```

```python
date
```

```python
#date.info()
```

```python
df1 = date
df1 = df1.copy()
```

```python
df1['time'] = pd.to_datetime(df1['time'])
```

```python
#df1.info()
```

```python
df1['date'] = df1['time'].dt.date
# df1['hour'] = df1['time'].dt.hour
# df1['time'] = df1['time'].dt.time
```

```python
df1
```

```python
result = df1.groupby(['date', 'lat', 'lon'])['PRES_L1'].mean().reset_index()
```

```python
result
```
