```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df= pd.read_excel("Combined All Stations_100m+PM25 added preci.xlsx")
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
# Filter data for dates between January 1, 2021, and December 31, 2022
date_mask = (df['time'] >= '2021-01-01') & (df['time'] <= '2022-11-30')
filtered_data = df[date_mask]

# Create a new DataFrame with the filtered data
new_df = pd.DataFrame(filtered_data)

# Convert 'Time' column to datetime format
new_df['time'] = pd.to_datetime(new_df['time'])

# Display the new DataFrame
print(new_df)
```

```python
# Remove rows with NaN values
new_df= new_df.dropna()
```

```python
# high_pm= df[df['PM 2.5']>30]
```

```python
# high_pm
```

```python
# Assuming your DataFrame is named 'df'
X = new_df.drop(['PM 2.5','FRICV_L1','LAND_L1','Station Name',"time","lat","lon","DLWRF_L1_Avg_1"], axis=1)  # Features (excluding the target variable)
y = new_df['PM 2.5']  # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
#print(X_test.dtypes)
```

```python

# Create the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)


```

```python
feature_importance = pd.Series(rf_model.feature_importances_, index=X.columns)
feature_importance.sort_values(ascending=False, inplace=True)
print(feature_importance)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Make predictions on the training set
y_train_pred = rf_model.predict(X_train)

# Make predictions on the test set
y_test_pred = rf_model.predict(X_test)

# Calculate RMSE for training and testing sets
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

# Calculate R-squared for training and testing sets
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Display the results
print(f'Training RMSE: {rmse_train:.4f}')
print(f'Testing RMSE: {rmse_test:.4f}')

print(f'Training R-squared: {r2_train:.4f}')
print(f'Testing R-squared: {r2_test:.4f}')

```

```python
import joblib

# Save the model to a file
model_filename = 'random_forest_model22.pkl'
joblib.dump(rf_model, model_filename)

```
