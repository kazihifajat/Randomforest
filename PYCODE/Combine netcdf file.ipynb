## Read all nc file and combine all into a single file

```python
# List all the .nc files in the directory
nc_files = glob.glob(r'D:\PM_DATA\UNZIPED_DATA\totalcloudcover\*.nc')

# Open all the .nc files using xarray
datasets = [xr.open_dataset(file) for file in nc_files]

# Combine the datasets along a new dimension (e.g., 'time')
combined_dataset = xr.concat(datasets, dim='time')

# Save the combined dataset to a new .nc file
combined_dataset.to_netcdf(r'D:\PM_DATA\EXTRACTED_NETCDF_DATA\tot_cloud_cover.nc')

# Close the datasets
for ds in datasets:
    ds.close()
```
