```python
date = df[(df['time'] >= '2015-02-01 00:00:00') & (df['time'] <= '2023-08-31 18:00:00')]
```

```python
df1 = date
df1 = df1.copy()
```

```python
df1['date'] = df1['time'].dt.date
# df1['hour'] = df1['time'].dt.hour
# df1['time'] = df1['time'].dt.time
```

```python
result = df1.groupby(['date', 'lat', 'lon'])['GUST_L1'].mean().reset_index()
```
